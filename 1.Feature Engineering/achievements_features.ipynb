```python
import pandas as pd
```

```python
achieves = pd.read_pickle('./achievements.pkl')
achieves.head()
```

```python
achieves['achieve_year'] = [int(value.split('/')[1])+2000 if len(value)!=4 else int(value) for value in achieves['Value']]
achieves.head()
```

```python
achieves['Transfer_Date'] = pd.to_datetime(achieves['Transfer_Date'], errors='coerce')
achieves['Transfer_Year'] = achieves['Transfer_Date'].dt.year
achieves['Transfer_Month'] = achieves['Transfer_Date'].dt.month
achieves.head()
```

```python
achieves['achieves_include_flag'] = 0

for idx, row in achieves.iterrows():
    if (row['Transfer_Month'] <=5) & (row['achieve_year']<row['Transfer_Year']):
        achieves.loc[idx,'achieves_include_flag'] = 1
    if (row['Transfer_Month'] > 5) & (row['achieve_year']<=row['Transfer_Year']):
        achieves.loc[idx,'achieves_include_flag'] = 1
achieves.head()
        
```

```python
achieves.head(50)
```

```python
achieves = achieves[achieves['achieves_include_flag'] ==1 ]
achieves.info()
```

```python
total_achievements = achieves[['Transfer_id','Name']].groupby('Transfer_id').count()
total_achievements.columns = ['Total_Achievements']
```

```python
distinct_achievements = achieves[['Transfer_id','Name']].drop_duplicates().groupby('Transfer_id').count()
distinct_achievements.columns = ['Distinct_Achievements']
```

```python
achieves_features = achieves[['Transfer_id']].drop_duplicates().set_index('Transfer_id')
achieves_features
```

```python
achieves_features = achieves_features.merge(total_achievements, left_index=True, right_index=True, how='left')
achieves_features = achieves_features.merge(distinct_achievements, left_index=True, right_index=True, how='left')
achieves_features
```

```python
achieves_features.reset_index().to_pickle('./achievement_features.pkl')
```
