```python
pos_groups = pd.read_csv('./position_group_national.csv', delimiter=";")
pos_groups.columns = ['position', 'Excuse']
```

```python
stats = stats.merge(pos_groups, on=['position'], how='left')
```

```python
stats
```

```python
#transfer id içerisinde gelen maç sayısı
stats.sort_values(by=['Transfer_id','match_date'],ascending=False, inplace=True)
stats['match_count'] = stats.groupby('Transfer_id').cumcount()
```

```python
stats['minutes_played'] = pd.to_numeric(stats['minutes_played'].str.replace('\'',''),  errors='coerce')
stats['substitutions_on'] = pd.to_numeric(stats['substitutions_on'].str.replace('\'',''),  errors='coerce')
stats['substitutions_off'] = pd.to_numeric(stats['substitutions_off'].str.replace('\'',''),  errors='coerce')
stats['goals'] = pd.to_numeric(stats['goals'], errors='coerce')
stats['assists'] = pd.to_numeric(stats['assists'], errors='coerce')
stats['own_goals'] = pd.to_numeric(stats['own_goals'], errors='coerce')
stats['yellow_card'] = pd.to_numeric(stats['yellow_card'].str.replace('\'',''), errors='coerce')
stats['second_yellow_card'] = pd.to_numeric(stats['second_yellow_card'].str.replace('\'',''), errors='coerce')
stats['red_card'] = pd.to_numeric(stats['red_card'].str.replace('\'',''), errors='coerce')
```
