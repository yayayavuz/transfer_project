```python
import requests, bs4,re
import pandas as pd
from datetime import timedelta, date
```

```python
def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)
```

```python
def scraper (start_date, end_date):
    df2=pd.DataFrame()
    for single_date in daterange(start_date, end_date):
        date2=str(single_date.strftime("%Y-%m-%d"))
        print(date2)
        url_first='https://www.transfermarkt.com/transfers/transfertagedetail/statistik/top/land_id_zu/0/land_id_ab/0/leihe//datum/'+ date2
        headers = {"User-Agent":"Mozilla/5.0"}
        response = requests.get(url_first, headers=headers)
        soup = bs4.BeautifulSoup(response.text, 'html.parser')
        last_page=soup.find_all(title=re.compile("Go the last page"))
        if not last_page:
            last_page_cnt=1
        else:
            last_page_cnt=int(re.findall(r'\d+',last_page[0].get('title'))[0])
        for x in range(last_page_cnt):
            print("Page"+str(x+1))
            url='https://www.transfermarkt.com/transfers/transfertagedetail/statistik/top/land_id_ab//land_id_zu//leihe//datum/'+date2+'/plus/0/galerie/0/page/'+str(x+1)
            response = requests.get(url, headers=headers)
            soup = bs4.BeautifulSoup(response.text, 'html.parser')
            table = soup.find_all('table')[1] 
            new_table = pd.DataFrame(columns=range(0,17),index=range(0,25))
            row_marker=0
            for row in table.find_all('tr'):
                columns = row.find_all('td')
               # print("row"+ str(row_marker))
                column_marker = 0
                for column in columns:
                    if row_marker % 7 == 1:
                        #print(str(int(row_marker/7))+","+str(column_marker))
                        #print("column"+str(column_marker))
                        #print(column.get_text().replace('\n',''))
                        new_table.iat[int(row_marker/7),column_marker] = column.get_text().replace('\n','')
                        if column_marker == 13:
                            new_table.iat[int(row_marker/7),14] = column.select("a")[0].get("href")
                            new_table.iat[int(row_marker/7),15] = column.select("a")[0].get("href").split('transfer_id')[1].replace('/','')
                        new_table.iat[int(row_marker/7),0] = row.select("a.spielprofil_tooltip")[0].get("id")
                        new_table.iat[int(row_marker/7),1] = row.select("a.spielprofil_tooltip")[0].get("href")
                    column_marker += 1
                row_marker =row_marker+1
            df = new_table.loc[:,[0,1,2,3,4,7,8,11,12,13,14,15]]
            df[16]=date2
            df[17]='Page'+' ' + str(x+1)
            df2=df2.append(df,ignore_index=True)
    return df2
```

```python
df1 = scraper(start_date = date(2016,1, 1),end_date = date(2017, 1,1 ))
```

```python
df2 = scraper(start_date = date(2017,1, 1),end_date = date(2018, 1,1 ))
```

```python
df3 = scraper(start_date = date(2018,1, 1),end_date = date(2019, 1,1 ))
```

```python
df4 = scraper(start_date = date(2019,1, 1),end_date = date(2020, 1,1 ))
```

```python
df5 = scraper(start_date = date(2020,1, 1),end_date = date(2020, 10,19 ))
```

```python
df1.to_pickle("./transfers_2016.pkl")
df2.to_pickle("./transfers_2017.pkl")
df3.to_pickle("./transfers_2018.pkl")
df4.to_pickle("./transfers_2019.pkl")
df5.to_pickle("./transfers_2020.pkl")
```

```python
frames = [df1,  df2,  df3, df4,df5]

df_all = pd.concat(frames)
```

```python
df_all.reset_index(drop=True, inplace=True)
```

```python
df_all=df_all.rename(columns={0:'id',1: 'Link',2:'Player_Name',3:'Player_Position',4:'Age',7:'Club_Left',8:'League_Left',11:'Club_Joined',12:'League_Joined',13:'Transfer_Fee',14:'Transfer_Link',15:'Transfer_id',16:'Transfer_Date',17:'Page_Number'})
cols=df_all.columns.tolist()
cols = cols[-4:] + cols[:-4]
df_all = df_all[cols] 
```

```python
df_all.to_pickle("./Transfers_2016_2020.pkl")
```
