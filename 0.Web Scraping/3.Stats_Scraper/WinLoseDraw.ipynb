```python
import pandas as pd
import numpy as np
```

```python
teams_played = pd.read_pickle('C:/Users/YAVUZ/Transfer Fee Project/Transfer Details/Team_played_All.pkl')
teams_played.head()
```

```python
all_matches = pd.read_pickle('C:/Users/YAVUZ/Transfer Fee Project/Stats_Scraper/Stats_Matches_All.pkl')
all_matches.info()
```

```python
stats = all_matches.merge(teams_played, on = ['Transfer_Link','Transfer_id','id','Transfer_Date'],
                         how = 'left')
```

```python
stats.info()
```

```python
stats['Date']=pd.to_datetime(stats['Date'],format="%b %d, %Y", errors='coerce')
stats['Transfer_Date']=pd.to_datetime(stats['Transfer_Date'],format="%b %d, %Y", errors='coerce')
stats.head()
```

```python
stats_df=stats[stats['Date']<stats['Transfer_Date']]
```

```python
stats_df.shape
```

```python
stats_df = stats_df.query('Date>=Start and Date<=End')
```

```python
stats_df.shape
```

```python
aa = stats_df['Result'].str.extractall('(\d+)')
aa = aa.unstack()
aa.columns = ['Home_Goals','Away_Goals']
stats_df = stats_df.merge(aa,how='inner', left_index=True, right_index=True)
stats_df.head()
```

```python
conditions = [(stats_df['Home_Team'] == stats_df['Team'])&(stats_df['Home_Goals']>stats_df['Away_Goals']),
             (stats_df['Home_Team'] == stats_df['Team'])&(stats_df['Home_Goals']==stats_df['Away_Goals']),
             (stats_df['Home_Team'] == stats_df['Team'])&(stats_df['Home_Goals']<stats_df['Away_Goals']),
             (stats_df['Away_Team'] == stats_df['Team'])&(stats_df['Away_Goals']>stats_df['Home_Goals']),
             (stats_df['Away_Team'] == stats_df['Team'])&(stats_df['Away_Goals']==stats_df['Home_Goals']),
             (stats_df['Away_Team'] == stats_df['Team'])&(stats_df['Away_Goals']<stats_df['Home_Goals'])]
outputs = ['win','draw','lose','win','draw','lose']
stats_df['Win_Draw_Lose'] = np.select(conditions, outputs)
```
