```python
def achievements_scrap(start,end):
    achieves=pd.DataFrame()
    for j in range(start,end):
        print(j)
        url2='https://www.transfermarkt.com/' + transfers['Link'][j].split('/')[1] +"/erfolge/" + transfers['Link'][j].split('/')[3]+"/"+  transfers['Link'][j].split('/')[4]
        headers = {"User-Agent":"Mozilla/5.0"}
        response = requests.get(url2, headers=headers)
        soup = bs4.BeautifulSoup(response.text, 'html.parser')
        row = 1
        kk = pd.DataFrame()
        for idx,ccy in enumerate(soup.find_all(class_="large-6 columns")):
            if idx >=1:
                for row2 in ccy.find_all('tr'):
                    kk.loc[row,'Name'] = ccy.find_all('span')[0].get_text()
                    kk.loc[row,'Value'] = row2.find_all('td')[0].get_text()
                    row+=1
        kk['Transfer_Link']=transfers_det['Transfer_Link'][j]
        kk['Transfer_id']=transfers_det['Transfer_id'][j]
        kk['id']=transfers_det['id'][j]
        kk['Transfer_Date']=transfers_det['Transfer_Date'][j]
        kk['Team_Left']=transfers_det['Team_Left'][j]
        kk['Team_Joined']=transfers_det['Team_Joined'][j]
        achieves = achieves.append(kk,ignore_index=True,sort=False)
    return achieves
```

```python
df3 = pd.DataFrame()
```

```python

for i in range(7728,9499,1):
    df2=achievements_scrap(i,i+1)
    df3 = df3.append(df2, ignore_index=True, sort=False)
```
